```python
import os
from zipfile import ZipFile, BadZipFile

extract_to_directory = "/blue/eel6935/share/dthiyagarajan/manifest-1694710246744_1/"

try:
    with ZipFile(zip_file_path, 'r') as zip_ref:
        for member in zip_ref.namelist():
            # Skip directories
            if member.endswith('/'):
                continue
            
            target_path = os.path.join(extract_to_directory, member)
            os.makedirs(os.path.dirname(target_path), exist_ok=True)

            with zip_ref.open(member) as source, open(target_path, "wb") as target:
                target.write(source.read())

        print(f"✅ All files extracted to: {extract_to_directory}")

except BadZipFile:
    print("❌ Error: The zip file is corrupted!")
except Exception as e:
    print(f"⚠️ Unexpected error: {e}")
```
